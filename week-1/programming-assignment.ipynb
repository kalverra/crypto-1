
Alright, what happens when we xor a space with a letter?

```
' ' = 0x20
0x20 ^ 0x41(A) = 0x61(a)
0x20 ^ 0x42(B) = 0x62(b)
0x20 ^ 0x43(C) = 0x63(c)
```

We also know that `E(A) ^ E(B) = A ^ B`.

`A[i] ^ B[i] = ' '` indicates that `A[i]` and `B[i]` are probably letters.

`A[i] ^ B[i] = letter` indicates that `A[i] = capital_letter|lower_case_letter` and `B[i] = ' '` (or vice versa).

Let's try and find out where spaces might be in our cipher plain texts. Most of the texts are going to be lower-case letters, so if we find that when we xor a cipher `A`, position `i` is frequently an upper case character, we can guess that `A[i]` is a space.

````python
possible_space_indexes = {}
for i in range(len(ciphertexts)):
  possible_space_indexes[i] = {}

for i in range(len(ciphertexts)):
  for j in range(len(ciphertexts)):
    if i == j:
      continue
    xor = strxor(hex_to_english(ciphertexts[i]), hex_to_english(ciphertexts[j]))
    for k in range(len(xor)):
      if xor[k].isupper():
        if k not in possible_space_indexes[i]:
          possible_space_indexes[i][k] = 1
        else:
          possible_space_indexes[i][k] += 1

for i in range(len(ciphertexts)):
  for index, count in possible_space_indexes[i].copy().items():
    if count >= 5: # Decent confidence this is a space
      cracked_ciphers[i][index] = " "
    else:
      del possible_space_indexes[i][index]

print_ciphers(cracked_ciphers)
````

We've probably got some false positives, but we can work with it. If we know `A[i] = ' '`, we can just flip the case of `A[i] ^ B[i]` and get the letter for `B[i]`.

````python
for i in range(len(ciphertexts)):
  for j in range(len(ciphertexts)):
    if i == j:
      continue
    xor = strxor(hex_to_english(ciphertexts[i]), hex_to_english(ciphertexts[j]))
    for space_index in possible_space_indexes[i].keys():
      if space_index >= len(cracked_ciphers[j]):
        break
      if (xor[space_index] >= 'a' and xor[space_index] <= 'z') or (xor[space_index] >= 'A' and xor[space_index] <= 'Z'):
        cracked_ciphers[j][space_index] = xor[space_index].swapcase()

print("'", strxor("xasdf", "asadfbbbbbasdfasf"), "'")
print_ciphers(cracked_ciphers)
````

We've got a little we can work with here, but not much. Let's throw some [crib dragging](https://samwho.dev/blog/toying-with-cryptography-crib-dragging/) at the problem. There are [some common English sequences](https://en.wikipedia.org/wiki/Most_common_words_in_English) like ` a ` and ` the `, and if we have `A ^ B`, we can try to eliminate either `A` or `B` from the equation by guessing that those sequences will be in one or the other. Let's focus on our target, xor it with all the other values, and drag some cribs through it and see if we get anything intelligible.

````python
english_sequences = ["The ", " the ", "A ", " a ", "And ", " and ", "In ", " in ", "But, ", ", but "]

for cipher in ciphertexts:
  xor = strxor(hex_to_english(cipher), hex_to_english(target))
  for seq in english_sequences:
    print("Crib: '", seq, "'")
    for i in range(len(xor) - len(seq)):
      drag = strxor(xor[i:i+len(seq)], seq)
      if all((c >= 'a' and c <= 'z') or (c >= 'A' and c <= 'Z') or c == ' ' for c in drag): # Narrow down to just letters and spaces
        print(i, i+len(seq), drag)
````

We've got a lot. Mostly gibberish, but there's some intelligible bits in there. Only issue is that in the case of `A ^ B ^ CRIB`, if the `CRIB` hits, then we can't be sure if we've eliminated `A` or `B`, so after some guess and check, and some inference, we've found the secret phrase, `The secret message is: When using a stream cipher, never use the key more than once`.
